## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill.manner_of_death.value_counts()
```

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='gender', y="age", data=kill, hue='manner_of_death')
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python

sns.swarmplot(x='gender', y="age", data=kill, hue='manner_of_death')
```

## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts().head(12)
```

```python
kill.city.value_counts().head(12).values
```

```python
kill.city.value_counts().head(12).index
```

```python
plt.figure(figsize=(8,4))
plt.title('Most dangerous cities',color='b')
sns.barplot(x=kill.city.value_counts().head(12).index,y=kill.city.value_counts().head(12).values,data=kill )
plt.xticks(rotation = 45);
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize=(8,4))
sns.countplot(x=kill.city, data= kill,order=kill.city.value_counts().head(12).index )
plt.xticks(rotation = 45);
```

```python
# with seaborn countplot
```
