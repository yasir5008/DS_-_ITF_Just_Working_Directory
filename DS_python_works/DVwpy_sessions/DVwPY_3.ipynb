```python
# Basic Libraries
import pandas as pd
import numpy as np
import scipy
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
```

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x='total_bill',y='tip',data=tips)
ax.set(xlabel='Total Bill', ylabel='Tip', title='Some title')
#plt.legend(labels=['legend1'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'total_bill', y= 'tip',data=tips,style='sex', hue='sex', s=200)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',s=200, data=tips, alpha=0.5);
#sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill',y='tip', data=tips,style='sex',hue='sex',s=100,alpha=0.7)
# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```

```python
sns.pairplot(tips, hue="sex")
```

```python
plt.figure(figsize=(14,10))
sns.rugplot(x='total_bill',data=tips,height=0.4)
#sns.rugplot(tips['total_bill'])
```

```python
sns.displot(y= 'total_bill',data=tips, rug=True)
```

```python
sns.histplot(tips.total_bill,kde=True, fill = False)
```

```python
plt.figure(figsize=(14,8))
sns.set(style="darkgrid")
sns.rugplot(data=tips, x="total_bill",height=0.1,color='r')
sns.histplot(data=tips, x='total_bill', kde=True, fill=False, bins= 30,color='b', lw = 4)
plt.show()
```

```python
sns.set(style='darkgrid', rc={"grid.color": ".6", "grid.linestyle": ":"})
sns.histplot(data = tips, x = 'total_bill',bins=20);
```

```python
sns.kdeplot(tips.total_bill)
```

```python
sns.kdeplot(tips.total_bill, clip=[10,20])
```

```python
sns.kdeplot(tips.total_bill, bw_adjust = 0.2,shade= True, color='r')
```
