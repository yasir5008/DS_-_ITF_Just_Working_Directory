```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
            diamonds.cut.isin(['Premium','Good']) &
            diamonds.color.isin(['D','F','J'])
        ].sample(n=100,random_state=22)

diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(diamonds.carat,diamonds.price);
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',data = diamonds);
```

```python
sns.scatterplot(x="x",y="y", data=diamonds)
```

```python
sns.scatterplot(x='carat',y='price',hue='depth',data = diamonds,);
```

```python
sns.scatterplot(x='carat',y='price',style='cut',data = diamonds, s=100);
```

```python
sns.scatterplot(x='carat',y='price',size='cut',data = diamonds);
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',style='color',data = diamonds, s=100);
```

```python
sns.scatterplot(x='carat',y='price',hue='depth',data = diamonds,s=200,alpha=0.6);
```

```python

sns.scatterplot(x='carat',y='price',hue='color',data = diamonds,s=200,alpha=0.6,marker='*');
```

```python
sns.scatterplot(x='carat',y='price',hue='color',data = diamonds,s=200,alpha=0.4,marker='*',edgecolor='black');
```
