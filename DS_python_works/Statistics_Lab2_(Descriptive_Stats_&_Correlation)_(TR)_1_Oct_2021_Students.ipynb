```python
tv_hours = np.array(tv_hours)
GPA = np.array(GPA)
```

Graph a Scatter Plot with Matplotlib

```python
plt.scatter(tv_hours,GPA)
```

Draw a fitting line

```python
m,b = np.polyfit(tv_hours,GPA,1)
plt.plot(tv_hours,GPA,'o')
plt.plot(tv_hours, m*tv_hours+b)
```
